# Spark Structured Streaming: Monolithic Structure

This notebook demonstrates Spark Structured Streaming with all logic in a single code block (monolithic). Compare this with the modular approach to observe any performance or execution plan differences.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import time

# All logic in one block
spark = SparkSession.builder \
    .appName("StreamingMonolithic") \
    .getOrCreate()
spark.sparkContext.setLogLevel("WARN")

# Data generation, transformation, aggregation, and streaming in one place
streaming_data = spark.readStream \
    .format("rate") \
    .option("rowsPerSecond", 100) \
    .option("rampUpTime", 1) \
    .load() \
    .withColumn("value", rand() * 100)

windowed_data = streaming_data \
    .groupBy(window("timestamp", "1 minute")) \
    .agg(avg("value").alias("avg_value"))

print("Execution Plan (Monolithic):")
windowed_data.explain(True)

def monitor_streaming_metrics(query):
    start_time = time.time()
    while query.isActive:
        progress = query.lastProgress
        if progress:
            print(f"\nBatch Duration: {progress.batchDuration} ms")
            print(f"Input Rows/sec: {progress.inputRowsPerSecond}")
            print(f"Processed Rows/sec: {progress.processedRowsPerSecond}")
            print(f"Memory Used: {progress.stateOperators[0].memoryUsedBytes if progress.stateOperators else 'N/A'} bytes")
        time.sleep(5)
        if time.time() - start_time > 60:
            query.stop()
            break

query = windowed_data.writeStream \
    .outputMode("complete") \
    .format("console") \
    .option("truncate", False) \
    .start()

monitor_streaming_metrics(query)
```

## Analysis

- All logic is in a single code block.
- Compare the execution plan and performance metrics with the modular notebook.
- In practice, Spark optimizes the logical plan, so performance differences are usually negligible for small/medium jobs, but modular code is easier to maintain and test.
